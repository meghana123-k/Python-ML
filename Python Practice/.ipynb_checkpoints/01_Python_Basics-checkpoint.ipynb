# Variables & Data Types

### Create variables of types int, float, string, bool, list, tuple, set, and dict. Print their type using type().

```python
intA = 45
floatB = 45.67
str = "Meghana"
isFake = True
lst1 = [34, 55, 78]
tple1 = (23, 45, 23)
set_nums = {45, 43, 43, 45}
dict = {"Name": "Meghana", "age": 20}
```

```python
print(type(intA))
print(type(floatB))
print(type(str))
print(type(isFake))
print(type(lst1))
print(type(tple1))
print(type(set_nums))
print(type(dict))
```

### Convert a string "123.45" into an integer and float. Handle the error if conversion fails.

```python
try:
    str = "123.45"
    n = int(float(str))
    print(n)
    num = int(str)
    print(num)
except Exception as e:
    print("Invalid..")
try:
    str = "123.45"
    num2 = float(str)
    print(num2)
except Exception as e:
    print("Invalid..")
```
